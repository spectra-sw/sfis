PRUEBA CODIFICACION IMAGENES

```python
from datetime import datetime, timedelta
minutes = timedelta(minutes=10)
now = datetime.now()
fmin =(now-minutes).strftime("%Y-%m-%d %H:%M:%S")
fmin = fmin[0:9]+"T"+fmin[11:]+"Z"
print(fmin)
```

```python
import cv2
import base64

cap = cv2.VideoCapture(0)
retval, image = cap.read()
retval, buffer = cv2.imencode('static/users/5a.jpg', image)
jpg_as_text = base64.b64encode(buffer)
print(jpg_as_text[:80])
cap.release()
```

CONVERSION DE BINARIO A IMAGEN

```python
jpg_original = base64.b64decode(jpg_as_text)

# Write to a file to show conversion worked
with open('test.jpg', 'wb') as f_output:
    f_output.write(jpg_original)
```
